```python
# Этот код конвертирует столбец 'opening_date_clean' в формат datetime и перезаписывает его в исходном DataFrame (df)
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

# Опредиляем отсутствующие значания и то как часто они встричаются

```python
# Команда дял определения отсутствующих значений или нулевых isna() Если значение пропущено, она возвращает значение True, в противном случае False
# sum(): эта функция вызывается после df.isna(), и ее цель - подсчитать количество True (пропущенных значений) для каждого столбца.
df.isna().sum()
```

# Определяем дублирующие значения

```python
# Полученный результат представляет собой датафрейм, который содержит только эти дублированные строки.
df.loc[df.duplicated()]
```

```python
#  В итоге, результатом выполнения строки кода будет новый датафрейм, содержащий только дублированные строки по указанному столбцу.
df.loc[df.duplicated(subset = ['Coaster_name'])]
```

```python
# Результатом выполнения данной строки кода будет новый датафрейм, содержащий только дублированные строки по указанным столбцам.
df = df.loc[df.duplicated(subset=['Coaster_name','Location','Opening_date_clean'])] \
   .reset_index(drop=True)
```
